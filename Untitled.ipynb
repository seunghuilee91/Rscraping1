```r
# 데이터 불러오기
txt <- readLines("hiphop.txt")
head(txt)
```

```r
# 특수문자 제거하기
# \\W는 특수문자를 의미하는 정규 표현식이다.
library(stringr)
txt <- str_replace_all(txt, "\\W", " ")
head(txt)
```

```r
# 명사 추출하기
noun <- extractNoun(txt)
```

```r
head(noun)
```

```r
# 추출한 명사 list를 문자열 벡터로 변환, 단어별 빈도표 작성
wordcount <- table(unlist(noun))
```

```r
wordcount
```

```r
# 데이터 프레임으로 전환
library(dplyr)
library(tidyverse)
df_wordcount <- as.data.frame(wordcount, stringsAsFactors = F)
df_wordcount <- rename(df_wordcount, word = Var1, freq = Freq)
# 빈도표 만들기
df_wordcount <- filter(df_wordcount, nchar(word) >= 2)
```

```r

# 상위 100개 추출하기
top_100 <- df_wordcount %>% 
 arrange(desc(freq)) %>% 
 head(100)
top_100
```

```r
# 워드 클라우드 만들기
library(wordcloud)
library(RColorBrewer)

# Rcolorbrewer의 함수를 통해 단어의 색깔을 지정할 때 사용할 색상 코드 목록을 만듭니다.
pal <- brewer.pal(8,"Dark2")

set.seed(1234)
wordcloud(words = df_wordcount$word, freq = df_wordcount$freq, max.words = 200, random.order = F, rot.per =.1, scale = c(4, 0.3), colors = pal)
```

```r


# Wordcloud2를 이용하는 방법
# 워드클라우드에 적용할 데이터의 길이를 최대 300건으로 제한합니다. 
# 이 숫자가 넘어가면 워드클라우드가 예쁘게 그려지지 않습니다. 
library(wordcloud2)
wordcloud2(top_100, color = "random-light", backgroundColor = "grey",fontFamily = 'HY견고딕')
wordcloud2(
 data = top_100,
 size = 0.8,
 fontFamily = 'HY견고딕',
 color = pal,
 backgroundColor = 'white',
 minRotation = -pi / 4,
 maxRotation = pi / 4,
 shuffle = TRUE,
 rotateRatio = 0.25,
 shape = 'circle',
 ellipticity = 0.6)
```

```r
# 고빈도 단어 트리맵을 그립니다. 
library(treemap)
treemap(
 dtf = top_100,
 title = 'TOP 100 DO IT R test',
 index = c('word'),
 vSize = 'freq',
 fontfamily.labels = 'HY견고딕',
 fontsize.labels = 14,
 palette = pal,
 border.col = 'white')
```
